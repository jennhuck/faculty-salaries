```python
averages = {}
```

```python
list_dept = {}
```

```python
for dept in departments:
    item = faculty.loc[faculty.dept==dept]
    list_dept[dept] = item
```

```python
list_dept
```

```python
for dept in departments:
    med = list_dept[dept].salary.median()
    std = list_dept[dept].salary.std()
    top_ratio = (list_dept[dept].salary.max())/med
    max = list_dept[dept].salary.max()
    averages[dept] = {"standard deviation" :std,"max":max, "med":med, "ratio": top_ratio}
```

```python
dist_data = pd.DataFrame(averages)
```

```python
dist_data = dist_data.transpose()
```
